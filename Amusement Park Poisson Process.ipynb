```python
import numpy as np
```

```python
hours = 12
rate = 1000
num_simulations = 1000

p_short = 0.24
p_med = 0.72
p_tall = 0.04
```

```python
tall_interarrivals = []
last_rider_type = []
extreme_tall_counts = 0
```

```python
for _ in range(num_simulations):
  # Get total arrivals
  total_arrivals = np.random.poisson(hours * rate)

  arrival_times = np.sort(np.random.uniform(0, hours, total_arrivals))

  # Height of arrivals based on given probability
  heights = np.random.choice(['short', 'medium', 'tall'], size=total_arrivals, p=[p_short, p_med, p_tall])

  # Tall arrival times
  tall_times = arrival_times[np.where(heights == 'tall')[0]]


  # Compute interarrival times of tall people
  if len(tall_times) > 0:
    interarrivals = np.diff(tall_times)
    tall_interarrivals.extend(interarrivals)

  # Check last rider type
  last_rider_type.append(heights[-1])

  # Check if at least 1000 tall people
  if np.sum(heights == 'short') >= 4000:
    extreme_tall_counts += 1
```

```python
avg_interarrival_tall_time = np.mean(tall_interarrivals) if tall_interarrivals else 0
```

```python
# Compute the probabilities for each type of rider being the last visitor
prob_last_short = last_rider_type.count('short') / num_simulations
prob_last_medium = last_rider_type.count('medium') / num_simulations
prob_last_tall = last_rider_type.count('tall') / num_simulations

# Probability of at least 1000 tall riders in a day
prob_extreme_tall_count = extreme_tall_counts / num_simulations
```
